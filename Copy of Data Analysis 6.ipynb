```python
df = df.drop("Unnamed: 0", axis=1)
print(df.head())
```

```python
df.head()
```

```python
# Gender Distribution
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df, x="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# From the above chart we have analysed that the number of females in the data is more than the number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb)
plt.show()
```

```python
# with values
plt.figure(figsize=(5,5))
sns.heatmap(gb,annot=True)
plt.title("Relationship Between Parent's Education and Student Score")
plt.show()
```

```python
# From the above chart we have conclude that the education of the parents have a good  impact score
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1,annot=True)
plt.title("Relationship Between Parent's Marital Status and Student Score")
plt.show()
```

```python
# From above figure we have conclude that there is no / neligible  impact
# on the student score due to their parent's martial status
```
